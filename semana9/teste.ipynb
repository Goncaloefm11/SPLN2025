```python
import spacy

# Carregar o melhor modelo gerado
nlp_ner = spacy.load("./output/model-best")

# Testar com uma frase em português
texto = "O Gonçalo estuda na Universidade do Minho e vive em Portugal."
doc = nlp_ner(texto)

# Ver o que ele encontrou
for ent in doc.ents:
    print(f"Entidade: {ent.text} | Etiqueta: {ent.label_}")
```
